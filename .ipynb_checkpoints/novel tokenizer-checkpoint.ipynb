```python
import requests
from bs4 import BeautifulSoup
import nltk
nltk.download('stopwords')
```

```python
r = requests.get('https://www.gutenberg.org/files/1342/1342-h/1342-h.htm') #pull pride and prejudice
```

```python
r.encoding = 'utf-8'

html = r.text

print(html[:2000])
```

```python
soup = BeautifulSoup(html, 'html.parser')

text = soup.get_text()

print(text[5000:8001])
```

```python
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')

tokens = tokenizer.tokenize(text)

print(tokens[:8])
```

```python
stop_words = nltk.corpus.stopwords.words('english')

print(stop_words)
```

```python
content_words = []
lowercase_words = []

for word in tokens:
    lowercase_words.append(word.lower())
    
print(lowercase_words[:8])
```

```python
for word in lowercase_words:
    if word not in stop_words:
        content_words.append(word)
        
print(content_words[:5])
freqdist = nltk.FreqDist(content_words)
print(freqdist)
```

```python


%matplotlib inline


freqdist = nltk.FreqDist(content_words)

freqdist.plot(25)
```
